```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Simple linear regression.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.drop('GPA', axis = 1)
y = df['GPA']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=2)
```

1. fit_intercept
  - Type: bool (default: True)
  - Description: Specifies whether to calculate the intercept (bias term) for the model.
    - If True, the model will learn the intercept.
    - If False, the intercept is set to 0.
  - Example: If your data is already centered (mean = 0), you can set this to False.
2. copy_X
  - Type: bool (default: True)
  - Description:
    - If True, a copy of the input data is made before fitting the model, ensuring the original data is not modified.
    - If False, the model may overwrite the input data (useful for saving memory).
  - Tip: Keep this True unless you're confident about memory constraints.
3. n_jobs
  - Type: int or None (default: None)
  - Description: The number of CPU cores to use for computation.
    - -1 means using all available cores.
    - None means using a single core.
  - Example: Useful when dealing with very large datasets.

```python
# Build linear Regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'MSE: {mse}')
print(f'MAE: {mae}')
```

```python
# apply polynomial
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.15, random_state=2)
```

```python
lr2 = LinearRegression()
lr2.fit(x_train, y_train)
y_pred = lr2.predict(x_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'MSE: {mse}')
print(f'MAE: {mae}')
```

```python
# apply lasso
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train, y_train)
```

```python
y_pred = lasso.predict(x_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'MSE: {mse}')
print(f'MAE: {mae}')
```

```python
# apply ridge
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train, y_train)
```
